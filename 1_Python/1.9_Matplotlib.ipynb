<h1 align='center'> Matplotlib </h1>

Matplotlib is a Python library that allow us to make several types of graphs as we will see in this notebook. We will be using the `pyplot` module in `matplotlib`, an we import this module as `plt` (recall to install the library first via `pip` or `conda`). 

First we import the necessary libraries:

```python
import matplotlib.pyplot as plt
import numpy as np
```

The `plot()` function graph the relationship between two variables, the first one we input will be the x-axis variable, and the second the y-axis variable.

```python
x = np.linspace(0, 100, 2000)
y = 5.0 + 0.8*x

plt.plot(x, y)
```

We could also plot several graphs in the same panel and assign to each one of them a different color and line type as follows:

```python
x = np.linspace(0, 20, 100)

plt.plot(x, 0.5*x**2, marker='.', color='cyan', label='plot 1')
plt.plot(x, 5*x, '--', color='green', label='plot 2')

plt.legend()
plt.show()
```

Here the `marker` argument means which type of line we want (dashed, dotted, solid line, etc) and the `label` argument assigns to each plot a name to be shown when using the `legend()` function. All the different possible markers can be found [here](https://matplotlib.org/stable/api/markers_api.html). And you can inspect to any color you want, either by typing the color name (e.g. red, blue, yelow), by introducing a hex code (e.g. #00FF00) or an RGB code (e.g. rgb(0, 255, 255)).

<h2> Subplots </h2

Sometime we will be interested in making several plots, for such a cases we would be wanting to place all the plots in a grid by using the `subplot()` function.

In the following example we place two different graphs in a grid that has two rows and one column, as specified in the `subplot()` function:

```python
x = np.linspace(0, 10, 100)

plt.figure()  # create a plot figure

plt.subplot(2, 1, 1) # (rows, columns, panel number
plt.plot(x, x**3,color="red")

plt.subplot(2, 1, 2)
plt.plot(x, 0.5*np.exp(x), color="orange");
```

Another way to achieve this is by assigning the figure and axis values returned from the `subplot()` function, which will define different positions for each plot and panel, so we can tell `matplotlib` which plot to place in which panel as in the next example:

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10))

# plot 1
x = np.linspace(-5, 5)
y = x**3

axs[0,0].plot(x, y, marker='>', color='#C24CF5')
axs[0,0].set_title('plot 1')

# plot 2
x = np.linspace(-500, 500)
y = np.log(x)

axs[0, 1].plot(x, y, marker='o', color = '#F5B07D')
axs[0, 1].set_title('plot 2')

# sine function
x = np.linspace(0,4*np.pi)
y = x*np.sin(x)

axs[1,0].plot(x, y, marker='d', color = '#F5D84C')
axs[1,0].set_title('plot 3')

# cosine function
x = np.linspace(0,4*np.pi)
y = x*np.cos(x)

axs[1,1].plot(x, y, marker='x', color = '#58E6F5')
axs[1,1].set_title('plot 4')
```

Here we use the `axs` variable to access each one of the panels in a (row, column) mode and the `figsize` argument in the `subplot()` function allow us to specify the width and height in inches of the complete grid.

<h2> Scatter plot </h2>

Scatterplots help us to inspect how two or more variables are related.

```python
import random

x = np.linspace(0, 100, 200)
random_nums = [random.randint(0, 5) for i in range(len(x))]

y = 10 + 0.5*x + random_nums

plt.scatter(x, y)
```

<h2> Bar plot </h2>

Barplots show us how a certain of numbers are assigned in different categories.

```python
# creating the dataset
courses = ['Economics', 'Medicine', 'Data science']
values = [10, 5, 20]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='aquamarine',
        width = 0.4)
 
plt.xlabel("Programs offered")
plt.ylabel("Number of students enrrolled")
plt.title("Amount of students enrolled in different academic programs")
plt.show()
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
eco = [12, 30, 1, 8, 22]
med = [28, 6, 16, 5, 10]
data_s = [29, 3, 24, 25, 17]
 
# Set position of bar on X axis
br1 = np.arange(len(eco))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
# Make the plot
plt.bar(br1, eco, color ='#34F48A', width = barWidth,
        edgecolor ='grey', label ='Economics')
plt.bar(br2, med, color ='#F54CC3', width = barWidth,
        edgecolor ='grey', label ='Medicine')
plt.bar(br3, data_s, color ='#40F2F5', width = barWidth,
        edgecolor ='grey', label ='Data Science')
 
# Adding Xticks
plt.xlabel('Year', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of students enrolled', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(eco))],
        ['2015', '2016', '2017', '2018', '2019'])

# Adding a title
plt.title('Number of students enrolled by program (2015-2019)')


plt.legend()
plt.show()
```

<h2> Pie plots </h2>

Pie plots show how a given variable is divided into several categories.

```python
labels = ['Approved', 'Deserted', 'Repeated']
sizes = [65, 20, 15]

fig, ax = plt.subplots()

plt.title('Distribution of students')

ax.pie(sizes, labels=labels, autopct='%1.1f%%',
      colors=['olivedrab', 'rosybrown', 'gray'])
```

<h2> Using matplotlib withing the Pandas library </h2> 

We can plot information that is contained in a `pandas` data frame by using the `plot()` method as wil be shown in the following examples. For that let's use the data used in the `pandas` section in the basic libraries notebook.

```python
import pandas as pd
```

```python
df = pd.read_csv('Data/gdp_per_capita.csv', encoding='latin1')
df.head()
```

```python
subset = df.head(10) # choose the first 10 countries
subset = subset[['Country', '2015']] # select variables of interest
subset.index = subset['Country'] # select variable to be used as an index
subset.drop('Country', axis=1, inplace=True) # drop column variable we assined as an index

# barplot
subset.plot(kind='bar', legend=False, title='GDP in 2015', xlabel='', ylabel='GDP')
```

The `kind` argument in the `plot()` method of a `pandas` dataframe asks which type of plot we want to make, common choices are:

- bar: vertical barplot
- barh: horizontal barplot
- box: boxplot
- line: lineplot
- hist: histogram
