Conversely, a continous random variable is a function that may take all the outputs in a given interval $[a, b]$. <h2> Density probability function </h2>

For a continous random variable, the density function is the homologous to the probability function in the discrete case. Let $f(x)$ be the density of the random variable $x$, then we define the probability of $x$ having values in the interval $[a, b]$ as:

$$P(a \leq x \leq b) = \int_{a}^{b} f(x)dx$$

Two understand the intuition behind this integral, think of it as computing the probability as every point in $[a, b]$ and then adding up, like it is shown in the following image:

<h2> Mathematical expectation and variance </h2>

To find the mathematical expectation of a continous random variable, we compute:

$$E[x] = \int_{- \infty}^{\infty} xf(x)dx$$

here the infinity bounds in the integral mean that we are integrating over all the domain of $f(x)$, such that we are averaging all the outcomes in the set of possible values of $f(x)$.

In the same way, the variance is computed as:

$$V[x] = E[(x - \bar{x})^2] = \int_{- \infty}^{\infty} (x - \bar{x})^2 f(x)dx$$

<h2> Important continous probability distributions </h2>

Although there are several continous distributions, here I will mention the most important distribution in data science and in statistics: the normal distribution.

<h3> Normal distribution </h2>

The normal distribution is by far the most known and used distribution in data science and in statistics in general. It is also known as Gaussian distribution, names after Carl Gauss. If $x$ is a random variable that follows a normal distribution, then it's probability function is

$$f(x) = \sqrt{\frac{1}{2 \pi \sigma}} e^{\frac{-(x - \bar{x})^2}{2 \sigma^2}}$$

where $\bar{x}$ is the mean, $\sigma$ is the standard deviation and $\pi$ is the irrational constant $3.14159...$

Now we plot how this function looks like fo three different combinations of mean and standard deviation. Note that the `scipy` library has a `norm` module that gives us functions to get the normal distribution, here we use the `pdf()` function that allow us to access the probability function density:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#x-axis ranges from -5 and 5 with .001 steps
x = np.arange(-5, 5, 0.001)

#define multiple normal distributions
plt.plot(x, norm.pdf(x, 0, 1), label='μ: 0, σ: 1')
plt.plot(x, norm.pdf(x, 0, 1.5), label='μ:0, σ: 1.5')
plt.plot(x, norm.pdf(x, 0, 2), label='μ:0, σ: 2')

#add legend to plot
plt.legend()
```
