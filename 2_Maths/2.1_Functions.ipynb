<h1 align='center'> Functions </h1>

In mathematical analysis we sometimes want to define the relationship among two or more phenomena like the temperature and the likelihood it will rain, or the amount of time I study for an exam and the score I get on the exam. To structure the definition of a given relationship, first we need to define the sets of possible values each variable can take, let's name them variables $x$ and $y$, and a given $(x, y)$ pair will be an observed realization of the underlying relationship between the two variables. For example, if $x$ is the number of hours I studied for my data science test, and $y$ the score I get in the test, then $(15, 4.0)$ means I studied 15 hours and got a 4.0 score. 

In the above example, it is normal to assume an positive relationship between $x$ and $y$; suppose that the relationship between these two variables is:

$$y = f(x) = ln(x)$$

where $y$ is the score I obtain in the exam, $x$ is the number of hours I study for the exam, $f(x)$ means "function of x", and $ln()$ is the natural-logarithm function. This relationship basically means that for each given number of hours I study for the exam, the receive score will be its natural logarithm.

Let's make a graph to see the underlying relationship:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 100)
y = [np.log(i) for i in x]

plt.plot(x, y)
plt.xlabel('Hours of study')
plt.ylabel('Test score')
plt.title('Relationship between the number of hours studied and test score')
```

If we want to be more realistic, we need to constrain the $x$ and $y$ variables in a given set, because one cannot study more than 24 hours a day, and tipically a test score is bounded in a given range. So, assume that $y \in [0, 5]$ and you have one week to prepare for the exam, so the maximum number of hours that you could theoretically study would be $24x7 = 168$, so the function would be given like:

 $$f(x) =\begin{cases}
 5 & \text{ if } x>148 \\ 
 ln(x) & \text{ if } 1 \leq x\leq148 \\ 
 0 & \text{ if } x<1
\end{cases}$$

```python
x = np.linspace(1, 100)
y = [np.log(i) for i in x]
x = np.append(x, np.linspace(148, 168))
y = y + [5 for i in np.linspace(148, 168)]

plt.plot(x, y)
plt.xlabel('Hours of study')
plt.ylabel('Test score')
plt.title('Relationship between the number of hours studied and test score')
```

<h2> Common functional forms </h2>

Once we have understood how to interpret a function $f(x)$ as a relationship between two variables, it is useful to examine the some common functions and their respective plots:

```python
fig, axs = plt.subplots(3, 2, figsize=(15,10))

# cuadratic function
x = np.linspace(-5, 5)
y = [i**2 for i in x]

axs[0,0].plot(x, y)
axs[0,0].set_title('$f(x) = x^2$')

# linear function
x = np.linspace(-500, 500)
y = [i*2 + 5 for i in x]

axs[0, 1].plot(x, y)
axs[0, 1].set_title('$f(x) = 2x + 5$')

# sine function
x = np.linspace(0,4*np.pi)
y = np.sin(x)

axs[1,0].plot(x, y)
axs[1,0].set_title('$f(x) = sin(x)$')

# cosine function
x = np.linspace(0,4*np.pi)
y = np.cos(x)

axs[1,1].plot(x, y)
axs[1,1].set_title('$f(x) = cos(x)$')

# absolute value function
x = np.arange(-500, 500, 1)
y = np.abs(x)

axs[2,0].plot(x, y)
axs[2,0].set_title('$f(x) = |x|$')

# exponential function
x = np.linspace(-5, 5)
y = np.abs(np.exp(x))

axs[2,1].plot(x, y)
axs[2,1].set_title('$f(x) = e(x)$')

for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')
    
fig.tight_layout(pad=2.0)
```

<h2> Multiple variables functions: a more realistic scenario </h2>

Believing that a given phenomenon might be related to one, and only one, variable is too idealistic and a little naive. For this reason, we might need to exend the aforementioned explanations to the case where we might have a function $f(x_{1}, x_{2}, ... , x_{n})$. For example, it would be feasible that the score we get on the data science exam would not solely depend on the amount of study we implement, but also on whether we find ourselves a little under the wheather, we have had good quality sleep and whether on the test day we are in a good mood. And also, in a real data science problem we will seldom have only two variables to relate, most serious problems have tenths, if not hundreads of variables.

All the mentioned (and plotted) functions in the last section can be generalized to a multiple variables scenario as follow:

- Cuadratic function: $f(x_{1}, x_{2}, ... , x_{n}) = x_{1}^{2} + x_{2}^{2} + ... + x_{n}^{2}$
- Linear function: $f(x_{1}, x_{2}, ... , x_{n}) = 5 + 2x_{1} + 2x_{2} + ... + 2x_{n}$
- Sine function: $f(x_{1}, x_{2}, ... , x_{n}) = sin(x_{1}) + sin(x_{2}) + ... + sin(x_{n})$
- Cosine function: $f(x_{1}, x_{2}, ... , x_{n}) = cos(x_{1}) + cos(x_{2}) + ... + cos(x_{n})$
- Absolute function: $f(x_{1}, x_{2}, ... , x_{n}) = |x_{1}| + |x_{2}| + ... + |x_{n}|$
- Exponential function: $f(x_{1}, x_{2}, ... , x_{n}) = e^{x_{1}} + e^{x_{2}} + ... + e^{x_{n}}$

From a geometrical viewpoint it is not possible to plot a function of more than two variable, thus I will plot a the cuadratic function $f(x_1, x_2) = x_{1}^{2} + x_{1}^{2}$ in a 3D plot, and I will leave the rest for you to plot:

```python
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')

X, Y = np.mgrid[-5:5, -5:5]
Z = X**2 + Y**2

ax.plot_surface(X, Y, Z)

plt.title('$f(x_1, x_2) = x_{1}^{2} + x_{1}^{2}$')

ax.set_xlabel('$x_{1}$')
ax.set_ylabel('$x_{2}$')
ax.set_zlabel('$y$')

plt.show()
```
