<h2> Introduction </h2>
In addition to understand how to variables are related, one might also be interested in how much a $y$ variable changes if $x$ changes. The computation of this is what we call a derivative, and this concept is what this notebook is about. Let's take a simple function like $f(x) = 4x + 2$, in this case it is easy to understand that if $x$ increases in a single unit, then $y$ will increase in 4 units. To see this, let's put a pin in a single point of the function, for instance $x=3$ which means that $y=14$; if suddenly $x=4$, then $y=18$, thus the increase in $y$ ($\Delta y$) is $4$ ($18-14$).

In many applications understanding how a variable will response when increasing another variable will have important implications in a given field. For example, in economics the derivative of the quantity demanded of a given product with respect to market prices gives important information regarding taxation and has implications in fiscal policy.

<h2> Mathematical notation of derivatives </h2>

Formally the derivative of $y$ with respect to $x$ is simply the computation of the change of $y$ when $x$ changes. Using mathematical notation we can write this like any of the following forms:

- $f'(x)$
- $\frac{dy}{dx}$
- $\frac{df(x)}{dx}$

The formal definition of a derivative is:

$$f'(x) = \frac{dy}{dx} = \frac{df(x)}{dy} = \lim_{\Delta x \rightarrow 0}\frac{\Delta f(x)}{\Delta x}$$

<h2> Geometric interpretation </h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10)
y = x**3 + x**2 + x

plt.plot(x, y)

tan1 = 66*x[0:30]+ 239
plt.plot(x[0:30], tan1, label='Tangent 1')

tan2 = 86*x[25:50] - 280
plt.plot(x[25:50], tan2, label='Tangent 2')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.plot(-5, -105, 'o')
plt.plot(5, 155, 'o')

plt.legend()

plt.title('Two derivatives of the function $f(x) = x^{3} + x^{2} + x$');
```

From a geometrical viewpoint let's consider the function $f(x) = x^{3} + x^{2} + x$ plotted above. If we want to find the derivative of the function when $x=-5$ (represented by the red dot in the graph) we will find the tangent line at this point, which is the orange line the the graph, meaning an straight line that "just" touches the function at that point, but do not cross it. Then we say the derivative of $f(x) = x^{3} + x^{2} + x$ at $x=-5$ is given by the slope of tangent 1. Note that in the graph I have plotted also a second tangent at $x=5$, and that tangent 1 and tangent 2 have different slopes (this is easy to see because both lines are not paralell), meaning that the concept of a derivative is a local concept rather than a global concept, this is, the derivative depends on which part of the function we are examining.

<h2> Derivative rules </h2>

Now let's introduce the general rules to compute the derivative of a function:

- $\frac{d(\alpha)}{x}=0$
- $\frac{d(\alpha x)}{dx} = \alpha$
- $\frac{d(x^{n})}{dx} = nx^{n-1}$
- $\frac{d(\alpha f(x))}{dx} = \alpha \frac{df(x)}{dx}$
- $\frac{d{f(x) + g(x)}}{dx} = \frac{df(x)}{dx} + \frac{dg(x)}{dx}$
- $\frac{d{f(x) - g(x)}}{dx} = \frac{df(x)}{dx} - \frac{dg(x)}{dx}$
- $\frac{d(f(x)g(x))}{dx} = f'(x)g(x) + f(x)g'(x)$
- $\frac{d(f(x)/g(x))}{dx} = \frac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}$

Let's use these rules to derivate the function $f(x) = x^{3} + x^{2} + x$ and find the values of tangent 1 and tangent 2 slopes:

$$f'(x) = 3x^{2} + 2x + 1$$

and

$$f'(-5) = 66$$
$$f'(5) = 86$$

Note that in most cases the derivative of a function will give another function at first from which one needs to replace a value to find the derivative at a given point. However, note that the derivative of a linear function will give always a constant value, implying that the derivative is the same regardless on which part of the function we are interested in.

<h2> Derivative of multiple variables functions</h2>

The only difference between the concept of derivating a single variable function and a multiple variable function is the fact that in the latter case we need to specify with respecto to which variable we are derivating. If we have a multiple variables function $f(x_{1},x_{2}, ..., x_{n})$ we can specify its derivativa with respect to $x_{1}$ using any of the following notations:

- $f_{1}(x_{1},x_{2}, ..., x_{n})$
- $\frac{\partial f(x_{1},x_{2}, ..., x_{n})}{\partial x_{1}}$

The only difference in the interpratation of this derivatives is that we now refer this to a partial derivative, meaning that we are calculatin the changes in $y$ when changes $x_{1}$ letting $x_{2}, x_{3}, ..., x_{n}$ constant.
